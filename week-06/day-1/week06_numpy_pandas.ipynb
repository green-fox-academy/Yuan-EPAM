```python
df_store = load_data(data_path1)
```

#### Preprocessing

```python
# Convert data with Rating value greater than 5 to None
df_store = df_store[(df_store['Rating'] <= 5) | (df_store['Rating'].isnull())]
```

```python
df_store['Reviews'] = df_store['Reviews'].astype(int)
```

```python
df_store['Rating'].shape
```

```python
def convert_size2MB(value):
    val = str(value)
    if 'Varies with device' == val:
        return None
    elif 'M' in val:
        if ',' in val:
            val = val.replace(',', '')
        return val.replace('M', '')
    elif 'k' in val:
        return float(val.replace('k', '')) / 1000
    else:
        return value

# Convert size to MB
df_store['Size'] = df_store['Size'].apply(convert_size2MB).astype(float)
```

#### What are the ten most recently updated apps?

#### Find apps, that is in the top 25% of apps based on ratings and were updted in the last two months of dataset

```python
rating_score = sorted(df_store['Rating'].dropna().unique())
rating_top25 = rating_score[math.floor((len(rating_score) * 0.75))]

# Top 25% apps based on ratings
df_store_rating_top25 = df_store[df_store['Rating'] > rating_top25]
```

```python
last_2_month = df_store.sort_values(by='Last_Updated_dt', 
                                    ascending = False)['Last_Updated_dt'].iloc[0] \
                                    - pd.DateOffset(months = 2)
last_2_month.to_period('M')
# Last 2 months (based on dataset date)
df_last_2_months = df_store[df_store['Last_Updated_dt'].dt.to_period('M') > last_2_month.to_period('M')]
```

```python
# Result 
np.intersect1d(df_store_rating_top25['App'], df_last_2_months['App'])
```
