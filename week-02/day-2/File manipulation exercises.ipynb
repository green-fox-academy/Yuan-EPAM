```python
def read_a_file(file_path):
    try:
        with open(file_path) as f:
            for line in f:
                print(line.rstrip())
    except IOError:
        print('Unable to read file: my-file.txt')
        
open_a_file('my-file.txt')
```

```python
def count_file_line(file_name):
    count = 0
    try:
        with open(file_name) as f:
            for line in f:
                count += 1
    except IOError:
        print(f'Unable to read file {file_name}')
    return count

print(count_file_line('test_file'))
print(count_file_line('test_file.txt'))
```

```python
def write_single_line(file_path, name):
    try:
        with open(file_path, 'a') as f:
            f.write(name)
    except IOError:
        print(f'Unable to read file {file_path}')
        
write_single_line('test_file.txt', 'Yuan')
read_a_file('test_file.txt')
```

```python
def write_multiple_lines(file_path, word, num):
    try:
        with open(file_path, 'a') as f:
            for i in range(num):
                f.write(word + '\n')
    except IOError:
        print(f'Unable to read file {file_path}')
   
file_path = './test_file.txt'
write_multiple_lines(file_path, 'apple', 5)
read_a_file(file_path)
```

```python
def copy_file(file_path_source, file_path_des):
    try:
        with open(file_path_source) as f_source:
            try:
                with open(file_path_des, 'w') as f_des:
                    for line in f_source:
                        f_des.write(line)
            except IOError:
                print(f'Unable to write the file {file_path_des}')
    except IOError:
        print(f'Unable to read file {file_path_source}')
         
file_path1 = 'test_file.txt'
file_path2 = 'test_file2.txt'
copy_file(file_path1, file_path2)
read_a_file(file_path2)
```

```python
def extract_IP(log_path):
    try:
        res = []
        with open(log_path) as f:
            for row, line in enumerate(f):
#                 print(line.split('  ')[2])
                res.append(line.split('  ')[2].strip(' /\n'))
        return res
    except IOError:
        print(f'Unable to read the file {log_path}')
        
logs_path = './log.txt'
extract_IP(logs_path)
```

```python
def get_request_ratio(log_path):
    try:
        res = {'GET': 0, 'POST': 0}
        with open(log_path) as f:
            for line in f:
                request = line.split('  ')[3].strip(' /\n')
                res[request] += 1
            if res['POST'] == 0:
                print('No POST request...')
                return 1
            return round(res['GET'] / res['POST'], 2)
    except IOError:
        print(f'Unable to read file {log_path}')
        
log_path = './log.txt'
get_request_ratio(log_path)
```

```python
def decrypt_doubled(file_path):
    try:
        res = []
        with open(file_path) as f:
            for line in f:
                res.append(line[::2])
        return res
    except IOError:
        print(f'Unable to read file {file_path}')
        
file_path = 'duplicated-chars.txt'
decrypt_doubled(file_path)
```

```python
def decrypt_reversed_lines(file_path):
    try:
        res = []
        with open(file_path) as f:
            for line in f:
                res.append(line[::-1].strip('\n') + '\n')
        return res
    except IOError:
        print(f'Unable to read the file {file_path}')
        
file_path = 'reversed-lines.txt'
decrypt_reversed_lines(file_path)
```

```python
def decrypt_reversed_order(file_path):
    try:
        res = []
        with open(file_path) as f:
            for line in f:
                res.append(line)
        return res[::-1]
    except IOError:
        print(f'Unable to read the file {file_path}')
        
file_path = 'reversed-order.txt'
decrypt_reversed_order(file_path)
```

```python
def decrypt_encoded(file_path):
    try:
        res = []
        with open(file_path) as f:
            for line in f:
                word_decoded = []
                for ch in line:
                    if ch != ' ':
                        word_decoded.append(chr(ord(ch) - 1))
                    else:
                        word_decoded.append(ch)
                res.append(''.join(word_decoded) + '\n')
        return res
    except IOError:
        print(f'Unable to read the file {file_Path}')
        
file_path = 'encoded-lines.txt'
decrypt_encoded(file_path)
```
